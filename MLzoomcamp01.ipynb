```python
q7 = islandData[['housing_median_age', 'total_rooms', 'total_bedrooms']]
q7
```

```python
x = q7.to_numpy()
x

xs = np.transpose(x)
xs
```

```python
XTX = x.dot(xs)
XTX
```

```python
y = np.array([950, 1300, 800, 1000, 1300])
y
```

```python
XTX_inv = np.linalg.inv(XTX)

XTX_inv
```

```python
wresult = xs.dot(XTX_inv)
wresult
```

```python
w = y * wresult
w
```
